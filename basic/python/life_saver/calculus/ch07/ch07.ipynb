**三角函数的极限**
> 1. 小数的情况:
>> 观察函数图像可以发现当 x 非常小的时候 sin(x) 非常接近与 x ，但是不能说 x = sin(x)。但是 $\lim\limits_{x\rightarrow 0}\frac{sin(x)}{x} = 1$ 可以说这个极限是成立的。而且这个极限式非常重要，这是解决三角函数微积分问题的关键所在。  
>>***
>> 再看 cos(x) ，cos(0) = 1 ，所以当 x 趋近于 0 时 cos(x) 趋近于 1 ，即 $\lim\limits_{x\rightarrow 0}cos(x) = 1$  
>>***
>> 再看 tan(x) ，$tan(x) = \frac{sin(x)}{cos(x)} \quad\Rightarrow\quad \lim\limits_{x\rightarrow 0}\frac{tan(x)}{x} = \lim\limits_{x\rightarrow 0}\frac{\frac{sin(x)}{cos(x)}}{x} = \lim\limits_{x\rightarrow 0}\big(\frac{sin(x)}{x}\big)\big(\frac{1}{cos(x)}\big) = (1)\big(\frac{1}{1}\big) = 1$   
>>***
>> 现在 $\lim\limits_{x\rightarrow 0}\frac{tan(x)}{x}$ 和 $\lim\limits_{x\rightarrow 0}\frac{sin(x)}{x}$ 都等于 1 ，那么$\lim\limits_{x\rightarrow 0}\frac{cos(x)}{x}$ 呢？$\lim\limits_{x\rightarrow 0^+}\frac{cos(x)}{x} = \infty$ ，$\lim\limits_{x\rightarrow 0^-}\frac{cos(x)}{x} = -\infty$ ，所以 $\lim\limits_{x\rightarrow 0}\frac{cos(x)}{x}$ $\color{green}{不存在，这点需要特别注意！！！！}$  
>>***
>> 总结:  
>> $\lim\limits_{x\rightarrow 0}\frac{sin(小数)}{小数} = 1$ ，$\lim\limits_{x\rightarrow 0}\frac{tan(小数)}{同样小的数} = 1$ ，$\lim\limits_{x\rightarrow 0}cos(小数) = 1$ ，这里前两个式子特别需要注意的是，分母和分子中正弦、正切的变量相匹配，如:  
>> $\lim\limits_{x\rightarrow 0}\frac{sin(x^2)}{x^2}$ 、 $\lim\limits_{x\rightarrow 0}\frac{sin(5x)}{5x}$ 、 $\lim\limits_{x\rightarrow 0}\frac{sin(3x^7)}{3x^7}$ 、 $\lim\limits_{x\rightarrow 0}\frac{sin(sin(x))}{sin(x)}$ 等  
>>***
>> 求 $\lim\limits_{x\rightarrow 0}\frac{sin(5x)}{x} = \lim\limits_{x\rightarrow 0}\frac{\frac{sin(5x)}{5x}\times 5x}{x} = \lim\limits_{x\rightarrow 0}\frac{sin(5x)}{5x} \times 5 = 1 \times 5 = 5$   
>>***
>> 求 $\lim\limits_{x\rightarrow 0}\frac{sin^3(2x)cos(5x^{19})}{xtan(5x^2)} = \lim\limits_{x\rightarrow 0}\frac{\big[\frac{(sin(2x))^3}{(2x)^3}\quad\times\quad (2x)^3\big]cos(5x^{19})}{x\big[\frac{tan(5x^2)}{5x^2}\quad\times\quad(5x)^2\big]} = \lim\limits_{x\rightarrow 0}\frac{\frac{(sin(2x))^3}{(2x)^3}\quad\cdot\quad cos(5x^{19})}{\frac{tan(5x^2)}{5x^2}}\quad\times\quad \frac{(2x)^3}{x(5x^2)} = \lim\limits_{x\rightarrow 0}\frac{\frac{(sin(2x))^3}{(2x)^3}\quad\cdot\quad cos(5x^{19})}{\frac{tan(5x^2)}{5x^2}}\quad\times\quad \frac{8x^3}{5x^3} = \frac{(1)^3(1)}{1} \times \frac{8}{5} = \frac{8}{5}$  
>>***
>> $\lim\limits_{x\rightarrow \infty}x sin(\frac{5}{x}) = \lim\limits_{x\rightarrow \infty}x \cdot \frac{sin(\frac{5}{x})}{\frac{5}{x}} \times \frac{5}{x} = \lim\limits_{x\rightarrow \infty}5 \times \frac{sin(\frac{5}{x})}{\frac{5}{x}} = 5 \times 1 = 5$  
>>***
>> 其他的三角函数可以先转化成 正弦、余弦、正切 后再计算即可。  
>> $\lim\limits_{x\rightarrow 0}sin(3x)cot(5x)sec(7x) = \lim\limits_{x\rightarrow 0}(sin(3x))\big(\frac{1}{tan(5x)}\big)\big(\frac{1}{cos(7x)}\big) = \lim\limits_{x\rightarrow 0}\big(\frac{sin(3x)}{3x}\times(3x)\big)\big(\frac{1}{\frac{tan(5x)}{5x}\times(5x)}\big)\big(\frac{1}{cos(7x)}\big) = (1)(\frac{1}{1})(\frac{1}{1})\times\frac{3}{5} = \frac{3}{5}$  
>>***
>> $\lim\limits_{x\rightarrow 0}\frac{1 - cos^2(x)}{x^2} \quad 由 sin^2(x) + cos^2(x) = 1 \quad\Rightarrow\quad \lim\limits_{x\rightarrow 0}\frac{1 - cos^2(x)}{x^2} = \lim\limits_{x\rightarrow 0}\frac{sin^2(x)}{x^2} = \lim\limits_{x\rightarrow 0}\big(\frac{sin(x)}{x}\big)^2 = 1$  
>>***
>> $\lim\limits_{x\rightarrow 0}\frac{1 - cos(x)}{x^2} = \lim\limits_{x\rightarrow 0}\frac{1 - cos(x)}{x^2}\times\frac{1 + cos(x)}{1 + cos(x)}  = \lim\limits_{x\rightarrow 0}\frac{1 - cos^2(x)}{x^2}\times\frac{1}{1 + cos(x)} = \lim\limits_{x\rightarrow 0}\big(\frac{sin(x)}{x}\big)^2\times\frac{1}{1 + cos(x)} = 1^2 \times \frac{1}{1 + 1} = \frac{1}{2}$  
>>***
>> $\lim\limits_{x\rightarrow 0}\frac{1 - cos(x)}{x} = \lim\limits_{x\rightarrow 0}\frac{1 - cos(x)}{x}\times\frac{1 + cos(x)}{1 + cos(x)} = \lim\limits_{x\rightarrow 0}\frac{1 - cos^2(x)}{x}\times\frac{1}{1 + cos(x)} = \lim\limits_{x\rightarrow 0}\frac{sin^2(x)}{x}\times\frac{1}{1 + cos(x)} = \lim\limits_{x\rightarrow 0}\big(sin(x) \times \frac{sin(x)}{x} \times \frac{1}{1 + cos(x)}\big) = 0 \times 1 \times \frac{1}{1 + 1} = 0$  
>>***
>> $\color{blue}{通过几何图形面积关系可以知道 sin(x) < x < tan(x) ， 0 < x < \frac{\pi}{2}}$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 10000)
y1 = x
y2 = np.sin(x)
y3 = np.tan(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
```

> 2. 大数的情况
>> 对于任意的 x 都有，$-1 \leq sin(x) \leq 1$ 以及 $-1 \leq cos(x) \leq 1$   
>> 求 $\lim\limits_{x\rightarrow \infty}\frac{sin(x)}{x}$ ，3.6 节中证明过 $\frac{sin(x)}{x}$ 是被夹在 $-\frac{1}{x}$ ， $\frac{1}{x}$ 之间的
所以根据$\color{red}{三明治定理}$ $\lim\limits_{x\rightarrow \infty}\frac{sin(x)}{x} = 0$  
>>***
>> 求 $\lim\limits_{x\rightarrow \infty}\frac{x sin(11x^7) - \frac{1}{2}}{2x^4}$ 其中 $-1 \leq sin(11x^7) \leq 1 \quad\Rightarrow\quad -x - \frac{1}{2} \leq x sin(11x^7) - \frac{1}{2} \leq x - \frac{1}{2}$ 又 $2x^4 > 0 \quad\Rightarrow\quad$ $\frac{-x - \frac{1}{2}}{2x^4} \leq \frac{x sin(11x^7) - \frac{1}{2}}{2x^4} \leq \frac{x - \frac{1}{2}}{2x^4}$   
>> 其中 $\lim\limits_{x\rightarrow \infty}\frac{-x - \frac{1}{2}}{2x^4} = \lim\limits_{x\rightarrow \infty}\frac{-1}{2x^3} - \lim\limits_{x\rightarrow \infty}\frac{1}{4x^4} = 0 - 0 = 0$   
>> 并且 $\lim\limits_{x\rightarrow \infty}\frac{x - \frac{1}{2}}{2x^4} = \lim\limits_{x\rightarrow \infty}\frac{1}{2x^3} - \lim\limits_{x\rightarrow \infty}\frac{1}{4x^4} = 0 - 0 = 0$  
>> 由 $\color{red}{三明治定理}$ 可知 $\lim\limits_{x\rightarrow \infty}\frac{x sin(11x^7) - \frac{1}{2}}{2x^4} = 0$  
>> $\color{blue}{常识: 对于任意的正指数\quad\alpha\quad\lim\limits{x\rightarrow\infty}\frac{sin(任何东西)}{x^\alpha}\quad= 0 \quad（注意对于余弦这仍然成立）}$  
>>***
>> $\lim\limits_{x\rightarrow 0}x sin(\frac{5}{x})$ $-1 \leq sin(\frac{5}{x}) \leq 1 \quad\Rightarrow\quad 当 x > 0 时 -x \leq x sin(\frac{5}{x}) \leq x ，\quad当 x < 0 时 x \leq x sin(\frac{5}{x}) \leq -x$ 于是分别计算左右极限 $\lim\limits_{x\rightarrow 0^+}x sin(\frac{5}{x}) = 0$ ，$\lim\limits_{x\rightarrow 0^-}x sin(\frac{5}{x}) = 0$ 所以 $\lim\limits_{x\rightarrow 0}x sin(\frac{5}{x}) = 0$

> 3. 其他的情况
>> $\lim\limits_{x\rightarrow\frac{\pi}{2}}\frac{cos(x)}{x - \frac{\pi}{2}}$ ，令 $t = x - \frac{\pi}{2}$ 则原式子变为$\lim\limits_{t\rightarrow 0}\frac{cos(t + \frac{\pi}{2})}{t}$ 又由三角恒等式 $cos(\frac{\pi}{2} - x) = sin(x) \quad\Rightarrow\quad cos(\frac{\pi}{2} + t) = -sin(t)$ 所以 $\lim\limits_{t\rightarrow 0}\frac{cos(t + \frac{\pi}{2})}{t} = \lim\limits_{t\rightarrow 0}\frac{-sin(t)}{t} = -1$  
>>***
>> $\color{blue}{总结: 面对 x\rightarrow a ，而 a \neq 0 时，有一个很好的一般原则，那就是用 t = x - a 做替换，将问题转化为 t\rightarrow 0 的问题}$

**三角函数的导数**
> y = sin(x) 的导数 $f^\prime(x) = \lim\limits_{h\rightarrow 0}\frac{sin(x + h) - sin(x)}{h}$ 由三角恒等式 $sin(A + B) = sin(A)cos(B) + cos(A)sin(B)\quad\Rightarrow\quad$ $f^\prime(x) = \lim\limits_{h\rightarrow 0}\frac{sin(x)cos(h) + cos(x)sin(h) - sin(x)}{h} = \lim\limits_{h\rightarrow 0}\frac{sin(x)(cos(h) - 1) + cos(x)sin(h)}{h} = \lim\limits_{h\rightarrow 0}\big(sin(x)\big(\frac{cos(h) - 1}{h}\big) + cos(x)\big(\frac{sin(h)}{h}\big)\big) = sin(x)\times 0 + cos(x)\times 1 = cos(x)$  
> y = cos(x) 时用同样的方法，由三角恒等式 $cos(A + B) = cos(A)cos(B) - sin(A)sin(B)$ 求出 $f^\prime(x) = -sin(x)$  
>***
> 其它三角函数的则可以用 $\color{red}{商法则、链式求导法则}$ 求出  
> $\frac{d}{dx}sin(x) = cos(x)$  
>  
> $\frac{d}{dx}cos(x) = -sin(x)$  
>  
> $\frac{d}{dx}tan(x) = -sec^2(x)$  
>  
> $\frac{d}{dx}sec(x) = sec(x)tan(x)$  
>  
> $\frac{d}{dx}csc(x) = -csc(x)cot(x)$  
>  
> $\frac{d}{dx}cot(x) = -csc^2(x)$  
>  
> $\frac{d^2}{dx^2}sin(x) = -sin(x)$  
>***
> 求 $\frac{d}{dx}(x^2 sin(x))$ 令 y = uv 其中 $u = x^2$ ，v = sin(x) ，则由乘法法则 $\Rightarrow \frac{dy}{dx} = v\frac{du}{x} + u\frac{dv}{x} = sin(x)\frac{d}{dx}(x^2) + x^2\frac{d}{x}(sin(x)) = 2x sin(x) + x^2 cos(x)$   
>***
> 求 $\frac{d}{dx}\big(\frac{sec(x)}{x^5}\big)$ 令 $y = \frac{u}{v}$ 其中 $u = sec(x)$ ，v = x^5 ，则由商法则 $\Rightarrow \frac{dy}{dx} = \frac{v\frac{du}{x} - u\frac{dv}{x}}{v^2} = \frac{x^5 sec(x)tan(x) - sec(x)\cdot 5x^4}{(x^5)^2} = \frac{sec(x)\cdot x^4(x tan(x) - 5)}{x^10} = \frac{sec(x)(x tan(x) - 5)}{x^6}$   
>***
> 求 $\frac{d}{dx}cot(x^3)$ 令 $u = x^3$ y = cot(u) 则由链式求导法则 $\Rightarrow \frac{dy}{dx} = \frac{dy}{du}\frac{du}{dx} = -csc^2(u)\cdot(3x^2) = -3x^2 csc^2(x^3)$   
>***
> $\color{blue}{常识: \frac{d}{dx}(sin(ax)) = a\cdot cos(ax) （如果用 ax 替换 x ，那么求导的时候前面会有一个额外的因子 a ）}$  
>***
> 三角函数自然而然会出现的一个地方是描述弹簧振子的运动。函数大致是 $x = 3sin(4t) 其中 x 是位移、t 是时间，且 3和4有可能会变，正弦也可能会变成余弦$ ，这种运动叫 $\color{red}{简谐运动}$

```python
x1 = np.linspace(-0.1, 0.1, 1000)
y1 = (x**2) * np.sin(1/x)

x = np.linspace(-0.1, 0.1, 1000)
y3 = x**2
y4 = -x**2

plt.plot(x1, y1)
plt.plot(x, y3)
plt.plot(x, y4)
```
