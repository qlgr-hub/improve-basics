**弧度:**  
> 旋转一周，我们说成 2$\pi$ 弧度而不是 $360^\circ$。因为半径为 1 的单位圆的周长是 $2\pi$ 个单位。这个圆的一个扇形的弧长就是这个扇形的圆心角的弧度。  
>***
> 用弧度度量的角 = $\frac{\pi}{180}\times$用度度量的角  
>***
> 学会用三角形来定义三角函数，假设我们有个直角三角形，除了直角以外的一个角被记为 $\theta$ 那么:  
> sin($\theta$) = $\frac{对边}{斜边}$ $\quad$ cos($\theta$) = $\frac{临边}{斜边}$ $\quad$ tan($\theta$) = $\frac{对边}{临边}$ $\quad$ 对边: 对着 $\theta$ 的边 $\quad$ 临边: 挨着 $\theta$ 的边 $\quad$ 斜边: 对着直角的边  
>***
> 倒数函数:$\quad$ csc(x) = $\frac{1}{sin(x)}$ $\quad$ sec(x) = $\frac{1}{cos(x)}$ $\quad$ tan(x) = $\frac{1}{tan(x)}$
>***
|$ $|$0$|$\frac{\pi}{6}$|$\frac{\pi}{4}$|$\frac{\pi}{3}$|$\frac{\pi}{2}$|
-|-|---------------|---------------|---------------|---------------|
|sin|$0$|$\frac{1}{2}$|$\frac{\sqrt 2}{2}$|$\frac{\sqrt 3}{2}$|$1$|
|cos|$1$|$\frac{\sqrt 3}{2}$|$\frac{\sqrt 2}{2}$|$\frac{1}{2}$|$0$|
|tan|$0$|$\frac{\sqrt 3}{3}$|$1$|$\sqrt 3$|$*$|
>***
> 角 $\theta$ 的参考角: 角 $\theta$ 与 x 轴之间的最小夹角，任意角的正弦就是其参考角正弦的正值或负值

```python
import numpy as np
import matplotlib.pyplot as plt
import math


x = np.arange(0, 2*math.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
```

```python
y = np.cos(x)
plt.plot(x, y)
```

```python
x = np.linspace(-np.pi / 2, np.pi / 2, 120)
y = np.tan(x)

plt.figure()
plt.plot(x, y)
plt.ylim([-10, 10])
ax = plt.gca()
ax.set_aspect(1)
plt.show()
```

```python
x = np.linspace(-np.pi / 2, np.pi / 2, 120)
y = 1 / np.cos(x)

plt.figure()
plt.plot(x, y)
plt.ylim([-10, 10])
ax = plt.gca()
ax.set_aspect(1)
plt.show()
```

```python
x = np.linspace(0.1, np.pi, 120)
y = 1 / np.sin(x)

plt.figure()
plt.plot(x, y)
plt.ylim([-10, 10])
ax = plt.gca()
ax.set_aspect(1)
plt.show()
```

```python
x = np.linspace(0.1, np.pi, 120)
y = 1 / np.tan(x)

plt.figure()
plt.plot(x, y)
plt.ylim([-10, 10])
ax = plt.gca()
ax.set_aspect(1)
plt.show()
```

**三角恒等式:**
> $\quad$ 1. $\quad$ tan(x) = $\frac{sin(x)}{cos(x)}$ $\quad$ cot(x) = $\frac{cos(x)}{sin(x)}$  
> $\quad$ 2. $\quad$ $cos^2(x) + sin^2(x) = 1$  
> $\quad$ 3. $\quad$ $1 + tan^2(x) = sec^2(x)$  
> $\quad$ 4. $\quad$ $cot^2(x) + 1 = csc^2(x)$
>***
> 两个角的合是 $\frac{\pi}{2}$ ，则称这俩角互余（complementary）  
> 三角函数(x) = co-三角函数$\left(\frac{\pi}{2} - x\right)$ , 如:  
> $\quad$ 5. $\quad$ sin(x) = cos$\left(\frac{\pi}{2} - x\right)$ $\quad$ tan(x) = cot$\left(\frac{\pi}{2} - x\right)$ $\quad$ sec(x) = csc$\left(\frac{\pi}{2} - x\right)$  
> $\quad$ 6. $\quad$ cos(x) = sin$\left(\frac{\pi}{2} - x\right)$ $\quad$ cat(x) = tan$\left(\frac{\pi}{2} - x\right)$ $\quad$ csc(x) = sec$\left(\frac{\pi}{2} - x\right)$
>***
> $\quad$ 7. $\quad$ sin(A + B) = sin(A)cos(B) + cos(A)sin(B) $\quad$ cos(A + B) = cos(A)cos(B) - sin(A)sin(B)  
> $\quad$ 8. $\quad$ sin(A - B) = sin(A)cos(B) - cos(A)sin(B) $\quad$ cos(A - B) = cos(A)cos(B) + sin(A)sin(B)  
> $\quad$ 9. $\quad$ sin(2x) = 2sin(x)cos(x) $\quad$ cos(2x) = $cos^2(x) - sin^2(x)$  
> $\quad$ 10. $\quad$ cos(2x) = $2cos^2(x) - 1$ = $1 - 2sin^2(x)$  
> 推导出:  
> $\quad$ sin(4x) = 2sin(2x)cos(2x) = $8sin(x)cos^3(x) - 4sin(x)cos(x)$ $\quad$ cos(4x) = $2cos^2(2x) - 1$ = $8cos^4(x) - 8cos^2(x) + 1$
