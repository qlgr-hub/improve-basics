**基本思想**
> 函数 f(x) 在区间 [a, b] 区间内画出图像，考虑该曲线、x 轴和两条垂直线 x = a 、x = b 围城图形的$\color{red}{有向面积}$可以记作 $\int_{a}^{b}f(x)dx$ 这就是$\color{red}{定积分}$。可以读作“函数 f(x) 对于 x 从 a 到 b 的积分”。  
>***
>$
\int_{0}^{1}xdx = 计算 两个直边都是 1 的三角形的面积 = \frac{1}{2}\times(1)\times(1) = \frac{1}{2} \\
\int_{0}^{2}xdx = 计算 两个直边都是 2 的三角形的面积 = \frac{1}{2}\times(2)\times(2) = 2 \\
\int_{-2}^{5}1dx = 计算 长为 7 宽为 1 的四边形的面积 = (1)\times(7) = 7 \quad\Rightarrow\quad \int_{a}^{b}1dx = b - a \quad\Rightarrow\quad \int_{a}^{b}dx = b - a \\
\\
如果 f 为奇函数 \int_{-a}^{a}f(x)dx = 0 （a 为任意实数）
$

**定积分的定义**
> 用按照面积定义的定积分如果遇到的不是三角形、四边形或偶函数这种简单场景就没法计算了，需要有相关的公式才可以的吧。  
>***
> $
\int_{a}^{b}f(x)dx = \lim\limits_{mesh\rightarrow 0}\sum\limits_{j=1}{n}f(c_j)(x_j - x_{j-1}) ，其中 a = x_0 < x_1 < ... < x_{n-1} < x_n = b ，并且对于每个 j = 1,...,n 都有 c_j 在 [x_{j-1}, x_j] 内。 \\
(1)\quad a = x_0 < x_1 < ... < x_{n-1} < x_n = b 告诉了我们，点 x_0,x_1,x_2,...,x_{n-1},x_n 形成了区间 [a, b] 的划分，其中最左边的 x_0 = a ，\\
\quad\quad最右边的 x_n = b 。这个划分创造了 n 个小的子区间 [x_0,x_1],[x_1,x_2],...,[x_{n-1},x_n] 。 \\
(2)\quad 划分中最大区间是指所有这些小区间中最长的区间，所以我们有，mesh = (x_1-x_0),(x_2-x_1),...,(x_{n-1}-x_{n-2}),(x_{n}-x_{n-1}) 中最长区间。 \\
(3)\quad 对于每个小区间，c_j 可以被选择在它所对应的区间的任何位置。这就是我们为什么说 c_j 在 [x_{j-1}, x_j] 内。 \\
(4)\quad 上述的极限是不断计算最大区间越来越少的不同的划分而求得的，也就是说，当它的最大区间趋于 0 时，n 趋于无穷。每个划分都涉及 c_j 的选择。 \\
(5)\quad 如果 f 是个连续函数，那我们怎样划分以及怎样选择 c_j 就显得无关紧要了，只要它的最大区间趋于 0 。事实上，只要函数 f 是有界的，即使 \\
\quad\quad它有有限个不连续点，这也是成立的。这样的函数是\color{red}{可积的}， 因为它可被积分。也有一些函数，即使它有无穷多个不连续点，也是可积的，但这已经 \\ 
\quad\quad超出了讨论的范围。另一方面，如果函数 f 是无界的，也可能是可积的，比如它有垂直渐近线，这种积分叫做反常积分。 \\
(6)\quad 在积分表达式中出现的求和 \sum\limits_{j=1}{n}f(c_j)(x_j - x_{j-1}) ，我们称之为\color{red}{利曼和}。它给出了定积分的估算值。如果它的最大区间非常的小，\\ \quad\quad那么这时估算将是非常精确的。
$  
>***
> 虽然这里有公式了，但是仍然很艰难其中做估算求和就很麻烦（需要等到微积分定理出来后才会有简单的方法去计算它！！！！！）

**定积分的性质**
> $
1. \int_{b}^{a}f(x)dx = -\int_{a}^{b}f(x)dx （汽车前进再往回走的例子很容易理解这个位移的符号的变化） \\
2. \int_{a}^{a}f(x)dx = 0 （物体运动分析根本没位移所以没面积） \\
3. \int_{a}^{b}f(x)dx = \int_{a}^{c}f(x)dx + \int_{c}^{b}f(x)dx （图像上很明显就是面积的和） \\
4. \int_{a}^{b}C f(x)dx = C \int_{a}^{b}f(x)dx （这里 C 是常数才可以成立的哟，只要转化成求和极限的形式一目了然） \\
5. \int_{a}^{b}(f(x) \pm g(x))dx = \int_{a}^{b}f(x)dx \pm \int_{a}^{b}g(x)dx （偷懒写一起了，文字描述就是和或差的积分等于积分的和或差） \\
$

**求面积**
> y = f(x) $\int_{a}^{b}f(x)dx$ 可以写成 $\int_{a}^{b}ydx$ ，这样就比较直观 y 就是拆分出的小长方形的高、 dx 就是它的宽，面积就是 ydx 。这个积分符号不仅仅是所有小长方形面积求和，它还要求每个长方形的宽 dx 趋于 0 。从这里可以发现积分可以用来计算面积。  
>***
> 1. 求通常的面积:  
>>$
通常直接用积分计算出的面积是有向面积（在 x 轴上面和在 x 轴下面符号是不一样的），要求通常面积其实就是不考虑面积的向 \\
求等于是给函数的绝对值求积分。求通常面积的过程: \\
(1)\quad 找出在 [a, b] 区间内满足函数值为 0 的所有 x 值。 \\
(2)\quad 以这些 x 值将 [a, b] 区间划分为多个子区间。 \\
(3)\quad 分别计算每个子区间的积分。 \\
(4)\quad 把计算出的所有子区间的积分挨个取绝对值，再相加。 \\
\\
对于运动物体: 位移 = \int_{a}^{b}v(t)dt \quad而，\quad 路程 = \int_{a}^{b}|v(t)|dt
$  
>***
> 2. 两条曲线之间的面积:  
>>$
函数 y = f(x) 、y = g(x) 、x = a 、x = b 所围成的面积: 在 f 和 g 之间的面积（平方单位） = \int_{a}^{b}|f(x) - g(x)|dx 。 \\
这里之所以用绝对值是因为我们无法保证两个函数的图像始终一个在另一个之上。 
$  
>***
> 3. 曲线与 y 轴所围成的面积:  
>>$
如果 f 存在反函数 ，\int_{A}^{B} f^{-1}(y)dy 就是由函数 y = f(x) 、直线 y = A 和 y = B 以及 y 轴所围成的面积（平方单位）。
$

**估算积分**
> $
\color{blue}{当一个函数一直都大于另一个函数时，它的积分也一直大于另一个函数的积分。} \\
即，如果对于在区间 [a, b] 内的所有 x 都有 f(x) \leq g(x) ，那么就有 \int_{a}^{b}f(x)dx \leq \int_{a}^{b}g(x)dx \\
\\
进而推导出一个简单的估算: \\
如果对于在 [a, b] 区间内的所有 x 有 m \leq f(x) \leq M ，那么 m(b - a) \leq \int_{a}^{b}f(x)dx \leq M(b - a) 
$

**积分的平均值和中值定理**
> 函数 f 在区间 [a, b] 内的平均值 = $\frac{1}{b - a}\int_{a}^{b}f(x)dx$ 。（从平均速度的求解中得到启发的，设 从 a 到 b 时间，时刻 t 的速度为 v(t) 则， 平均速度 = $\frac{位移}{总时间} = \frac{1}{b - a}\int_{a}^{b}v(t)dt$）  
>***
> $\color{red}{积分中值定理}$ : 如果函数 f 在闭区间 [a, b] 上连续，那么在开区间 (a, b) 内总有一点 c ，满足 $f(c) = \frac{1}{b - a}\int_{a}^{b}f(x)dx$ $\color{blue}{简而言之，连续函数在一段区间内至少一次达到它的平均值}$ 。  
>***
> 学微分时【11.3】节的中值定理是用 位移-时间 的图像中的斜率来解释的，而这里的中值定理则是用 速度-时间 图像中的面积来解释的。然后从结论上可以看出是一样的，都是在某个时间段内某一时刻的瞬时速度 = 这个时间段内的平均速度。  

**不可积函数**
>$
f(x) = 
\begin{cases}
1 \quad\quad 如果 x 是有理数 \\
2 \quad\quad 如果 x 是无理数
\end{cases}
$ $\quad\quad$，它就是个不可积函数，因为最大区间趋于 0 时 利曼上和与利曼下和不相等。  
> 其实有方法可以求出不可积函数的积分，叫$\color{red}{勒贝格积分}$，它与我们常用（正在用的）的$\color{red}{利曼和积分}$是相对的。
