## 1. 函数
函数是将一个对象转化为另一个对象的规则。起始对象称为输入，来自称为定义域的集合。返回对象称为输出，来自称为上域的集合。  
上域 vs. 值域: 上域是可能输出的集合，值域是实际输出的集合（值域是上域的子集，两个是不同的）
> 1. f(x) = x 表达中称 f(x) 为函数是错误的，应该 f 是函数，f(x) 是函数 f 在 x 上作用的结果  
> 2. 函数只有对在其定义域中的值作用才会有有效的输出  
> 3. 一个函数必须给每一个有效的输入指定唯一的输出  


```python
def f(x):
    return x**2

print(f(2))
print(f(-1/2))
print(f(1))
```

**区间：**
> 闭区间 [a,b]  
> 开区间 (a,b)  
> 半开区间 (a,b] 或 [a, b)  
> a 到 b 区间内不包含 c 的表示 [a,b]\{c}

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-2,1,0.1)
y1= x**2

plt.plot(x, y1, c='red')
```

**反函数：**
> y = f(x) 的反函数表示为 $f^{-1}$(y) = x  
> f(x) 反函数是 $f^{-1}$(x) ， $f^{-1}$(x) 的反函数是 f(x)  
> f(x) 和反函数 $f^{-1}$(x) 互相是镜子 y = x 的反射
> f(x) 的反函数的反函数是 f(x) 本身

**复合函数:**
> f(x) = h(g(x)) => f = h $\circ$ g

**奇函数和偶函数:**
> f(x) 在其定义域中所有的 x 满足 f(-x) = f(x)  => f(x) 是偶函数。（偶函数图像关于 y 轴有镜面对称性）  
> f(x) 在其定义域中所有的 x 满足 f(-x) = -f(x) => f(x) 是奇函数。（奇函数图像关于原点有180$^\circ$对称性）  
> 一般函数要么是偶函数，要么是奇函数，要么是非奇非偶函数，就有一个特例: f(x) = 0 是即奇又偶的函数  
***
> 两个奇函数的乘积是偶函数，两个偶函数的乘积是偶函数，一个奇函数和一个偶函数的乘积是奇函数:
> f(-x) = -f(x), g(-x) = -g(x) $\to$ h(x) = f(x) $\times$ g(x) $\to$ h(-x) = f(-x) $\times$ g(-x) = -f(x) $\times$ -g(x) = f(x) $\times$ g(x) = h(x)  
> f(-x) = f(x), g(-x) = -g(x) $\to$ h(x) = f(x) $\times$ g(x) $\to$ h(-x) = f(-x) $\times$ g(-x) = f(x) $\times$ -g(x) = -(f(x) $\times$ g(x)) = -h(x)  
> f(-x) = f(x), g(-x) = g(x) $\to$ h(x) = f(x) $\times$ g(x) $\to$ h(-x) = f(-x) $\times$ g(-x) = f(x) $\times$ g(x) = h(x)

**线性函数图像:**
> 对于通过 ($x_0$, $y_0$) 点斜率是 m 的直线由: y - $y_0$ = m(x - $x_0$)  
> 对于通过 ($x_1$, $y_1$) 和 ($x_2$, $y_2$) 点的直线由: (直线的斜率) m = $\cfrac{y_2 - y_1}{x_2 - x_1}$

**多项式:**
> p(x) = $a_nx^n + a_{n-1}x^{n-1} + \cdot\cdot\cdot + a_2x^2 + a_1x^1 + a_0x^0$ , 这里最高幂指数 n 叫做多项式的次数，最高次数项的系数 $a_n$ 叫首项系数    
> 多项式函数中多项式的次数决定函数的奇偶性，首项系数的正负决定函数图像的左右两端的走势，图像的中间部分由其他项决定  
>***
> 二次函数: p(x) = $ax^2 + bx + c$ , 由判别式: $\Delta = b^2 - 4ac$ 可以判断其实数解的情况
> 1. $\Delta$ > 0 ，有两个不同的实数解: $\cfrac{-b - \sqrt {b^2 - 4ac}}{2a}$ 和 $\cfrac{-b + \sqrt {b^2 - 4ac}}{2a}$  
> 2. $\Delta$ = 0 ，有一个实数解: $\cfrac{-b}{2a}$ ，即（$\cfrac{-b - \sqrt 0}{2a}$ = $\cfrac{-b + \sqrt 0}{2a}$）  
> 3. $\Delta$ < 0 ，无实数解
>***
> 二次函数的配方: $a^2 + 2ab + b^2$ = $(a + b)^2$ , $a^2 - 2ab + b^2$ = $(a - b)^2$
> $2x^2 - 3x +10$   
> $\to$ $2(x^2 - \cfrac{3}{2} + 5)$   
> $\to$ $2(x^2 - \cfrac{3}{2}x + 5 + \cfrac{9}{16} - \cfrac{9}{16})$  
> $\to$ $2((x^2 - \cfrac{3}{2}x - \cfrac{9}{16}) + 5 + \cfrac{9}{16})$  
> $\to$ $2((x - \cfrac{3}{4})^2 + 5 + \cfrac{9}{16})$   
> $\to$ $2(x - \cfrac{3}{4})^2 + \cfrac{71}{8})$

有理函数:

指数函数和对数函数:
> 对于指数函数 x 轴是它的水平渐近线  
> 对于对数函数 y 轴是它的水平渐近线  
> 指数函数与对数函数互为反函数

```python
x = np.arange(-10, 10, 0.1)

y = 2**x
plt.plot(x, y)
```

```python
y = 2**(-x)
plt.plot(x, y)
```

```python
y = (1/2)**x   # 就等于 2**(-x)
plt.plot(x, y) 
```

```python
x = np.arange(0.1, 10, 0.1)
y = np.log2(x)
plt.plot(x, y)
```

带有绝对值的函数:
> |x| = 
$\begin{cases}
x  \quad 如果 x \ge 0 \\
-x \quad 如果 x \lt 0 \\
\end{cases}$  
>***
> |x - y| 是数轴上 x 和 y 两点间的距离
>***
> 对于 y = x 的 x 轴下方部分以 x 轴为镜子映射到 x 轴上方就可以得到 y = |x| 的图像  
> 对 y = |$log_2x$| 也是如此

```python
x = np.arange(-10, 10, 0.1)
y = abs(x)
plt.plot(x, y)
```

```python
x = np.arange(0.1, 10, 0.1)
y = abs(np.log2(x))
plt.plot(x, y)
```
