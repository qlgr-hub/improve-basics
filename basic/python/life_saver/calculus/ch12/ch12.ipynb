# 绘制函数图像
> 1. 建立符号表格这种有用的技巧  
> 2. 绘制函数图像的一般方法  
> 3. 如何应用该方法的五个例子

**建立符号表格**
>***
> 1. 对函数本身做一个符号表格:  
>> $
（1） 选取 x : 统计函数的所有\color{red}{零点}和\color{red}{不连续点}，然后再在定义域中被这些点分成的区间内随机选取一些值。 \\
（2） 求出 f(x) 相关的符号: 对于所有选取的 x 求出 f(x) 。零点自然是 0 ，不连续点用 * 代替 ，对于那些随机选取的 x \\
\quad\quad只要知道对应 f(x) 的符号即可。 \\
（3） 对 选取的 x 和 求出的 f(x) 相关的信息 做表格（三行的表格，分别对应x、f(x)、保留行）
$  
>***
> 2. 对函数的一阶导数做一个符号表格:  
>> $
（1） 选取 x : 统计函数一阶导数的所有\color{red}{零点}和\color{red}{不连续点}（其实对于一阶导数这些是临界点），\\
然后再在定义域中被这些点分成的区间内随机选取一些值。 \\
（2） 求出 f^\prime(x) 相关的符号: 对于所有选取的 x 求出 f^\prime(x) 。零点自然是 0 ，不连续点用 * 代替 ，对于那些随机选取的 x \\
\quad\quad只要知道对应 f^\prime(x) 的符号即可。 \\
（3） 对 选取的 x 和 求出的 f^\prime(x) 相关的信息 做表格（三行的表格，分别对应x、f(x)、保留行: 一阶导数时这里用水平线或斜线表示其符号） \\
\color{blue}{用符号表示导数的正负: 一阶导数为正用向上的斜线表示，一阶导数为负用向下的斜线表示，一阶导数为零用水平线表示}
$  
>***
> 3. 对函数的二阶导数做一个符号表格:  
>> $
（1） 选取 x : 统计函数二阶导数的所有\color{red}{零点}和\color{red}{不连续点}，然后再在定义域中被这些点分成的区间内随机选取一些值。 \\
（2） 求出 f^\prime\prime(x) 相关的符号: 对于所有选取的 x 求出 f^{\prime\prime}(x) 。零点自然是 0 ，不连续点用 * 代替 ，对于那些随机选取的 x \\
\quad\quad只要知道对应 f^{\prime\prime}(x) 的符号即可。 \\
（3） 对 选取的 x 和 求出的 f^{\prime\prime}(x) 相关的信息 做表格（三行的表格，分别对应x、f(x)、保留行: 对于二阶导数用曲线符号表示凹形） \\
\color{blue}{符号表示凹形: f^{\prime\prime}(x) > 0 凹向上图形，f^{\prime\prime}(x) < 0 凹向下图形，f^{\prime\prime}(x) = 0 画个点}
$

**绘制函数图像的全面方法**
> 先准备好坐标轴  
>***
> $
1. 对称性: x = -x 代入先确定函数的奇偶性。因为奇函数或偶函数只要画 x \geq 0部分另一部分就根据对称性画出就性。 \\
2. y 轴截距: 通过 x = 0 求出 y 轴截距并标记到坐标轴上 \\
3. x 轴截距: 通过 y = 0 求出 x 轴截距并标记到坐标轴上。（对于有些函数这个可能有点难） \\
4. 定义域: 如果函数定义中没给出定义域时自己求出。（最好牢记常用的那些函数的定义域） \\
5. 垂直渐近线: 需要用虚线在坐标轴上标记处垂直渐近线。(有分母的函数通常出现在分母为 0 的点，或者对数因式也有可能出现) \\
6. 函数的正负: 根据上一节的方法，建立函数的符号表格来记录。 \\
7. 水平渐近线: 也是用虚线标记到坐标轴，通过求 \lim\limits_{x\rightarrow\infty}f(x) 或 \lim\limits_{x\rightarrow-\infty}f(x) 确定。（有渐近线时最好在做函数符号表格时选取一些渐近线附近的点确定图像在渐近线的哪一侧） \\
8. 一阶导数的正负: 根据上一节的方法，建立一阶导数的符号表格来记录。 \\
9. 最大值和最小值: 从一阶导数的符号表格求出所有的局部最大值和局部最大值标记到坐标轴上。 \\
10. 二阶导数的正负: 根据上一节的方法，建立二阶导数的符号表格来记录。 \\
11. 拐点: 根据一阶、二阶导数的符号表格找出所有拐点追加到函数符号表格中，同时标记到坐标轴上。
$
