# 极限导论
> 1. 对于极限是什么的直观概念
> 2. 左、右与双侧极限，及在 $\infty$ 和 $-\infty$ 处的极限
> 3. 何时极限不存在
> 4. 三明治定理（也称作“夹逼定理”）



**极限:基本思想**
> $\lim\limits_{x\rightarrow 2}f(x) = 1$ $\quad$ $f(x) \rightarrow 1 \quad 当 \quad x \rightarrow 2$ 
>***
>即便函数定义如下: 
$g(x) = 
\begin{cases}
x - 1,\quad x \neq 2\\
3, \quad\quad x = 2
\end{cases}$
> $\quad$极限仍然是: $\quad\lim\limits_{x\rightarrow 2}g(x) = 1$

**左极限与右极限**
> 对于函数
$h(x) = 
\begin{cases}
1,\quad x \lt 3\\
-2, \quad x \gt 3
\end{cases}$  
> 左极限: $\lim\limits_{x\rightarrow 3^-}h(x) = 1$，右极限: $\lim\limits_{x\rightarrow 3^+}h(x) = -2$
>***
> 只有左极限和右极限都存在，且相等时双侧极限才存在，否则双侧极限不存在（DNE）
>***
> 1. $f(x) = \frac{1}{x}$，当 x 趋近于 0 时:  
> 左极限是: $\lim\limits_{x\rightarrow 0^-}\frac{1}{x} = -\infty$  
> 右极限是: $\lim\limits_{x\rightarrow 0^+}\frac{1}{x} = \infty$  
> 2. $f(x) = \frac{1}{x^2}$，当 x 趋近于 0 时:  
> 左极限是: $\lim\limits_{x\rightarrow 0^-}\frac{1}{x^2} = \infty$  
> 右极限是: $\lim\limits_{x\rightarrow 0^+}\frac{1}{x^2} = \infty$  
> 所以双侧极限: $\lim\limits_{x\rightarrow 0}\frac{1}{x^2} = \infty$
>***
> 垂直渐近线: “f 在 x = a 处有一条垂直渐近线” 说的是 $\lim\limits_{x\rightarrow a^+}f(x)$ 和 $\lim\limits_{x\rightarrow a^-}f(x)$，其中至少有一个极限是 $\infty$ 或 $-\infty$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.00001, 48 * np.pi, 120000)
y = np.sin(1/x)

x1 = np.linspace(-0.00001, -48 * np.pi, 120000)
y1 = np.sin(1/x1)

plt.figure()
plt.plot(x, y)
plt.plot(x1, y1)
plt.ylim([-5, 5])
ax = plt.gca()
ax.set_aspect(48)
plt.show()
```

**在 $\infty$ 和 $-\infty$ 处的极限**
> “f 在 y = L 处有一条右侧水平渐近线”意味着 $\lim\limits_{x\rightarrow \infty}f(x) = L$  
> “f 在 y = M 处有一条左侧水平渐近线”意味着 $\lim\limits_{x\rightarrow -\infty}f(x) = M$  
> 如果一个数的绝对值是非常大的数，则这个数是大的  
> 如果一个数非常接近与 0（但不是真的等于 0），则这个数是小的  

```python
x = np.linspace(0.001, 4 * np.pi, 1200)
y = np.sin(x) / x

x1 = np.linspace(0.001, 4 * np.pi, 1200)
y1 = -1/x

x2 = np.linspace(0.001, 4 * np.pi, 1200)
y2 = 1/x

plt.figure()
plt.plot(x, y)
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.ylim([-1, 1])
ax = plt.gca()
ax.set_aspect(1)
plt.show()
```

**三明治定理**
> 如果一个函数 f 被夹在函数 g 和 h 之间，当 x $\rightarrow$ a 时，这两个函数 g 和 h 都收敛与同一个极限 L，那么当 x $\rightarrow$ a 时，f 也收敛与极限L。  
> g(x) $\leq$ f(x) $\leq$ h(x)，即 f(x) 被夹在 g(x) 和 h(x) 之间。此外，我们假设 $\lim\limits_{x\rightarrow a}g(x) = L$ 且 $\lim\limits_{x\rightarrow a}h(x) = L$。那么我们可以得出结论: $\lim\limits_{x\rightarrow a}f(x) = L$；即当 x $\rightarrow$ a 时，所有三个函数都有相同的极限。

```python
x = np.linspace(0.00001, 48 * np.pi, 1200000)
y = np.sin(x) / x

plt.figure()
plt.plot(x, y)
plt.ylim([-10, 10])
ax = plt.gca()
ax.set_aspect(5)
plt.show()
```

```python
x = np.linspace(0.00001, 120, 1200000)
y = x * np.sin(1/x)

plt.figure()
plt.plot(x, y)
plt.ylim([-10, 10])
ax = plt.gca()
ax.set_aspect(5)
plt.show()
```
