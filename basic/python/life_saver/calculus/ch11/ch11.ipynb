**函数的极值**
> 函数极值分为 全局极值 和 局部极值:  
>> 如果 x = a 时，f(a) 是函数 f 的整个定义域内的最大值，我们就说它是$\color{red}{全局最大值（绝对最大值）}$   
>>
>> 相对应的，如果 x = a 时，f(a) 是函数 f 的一个限定的区域（刨除定义域的大部分只留 a 附近的一小范围）内的最大值，我们就说它是$\color{red}{局部最大值（相对最大值）}$   
>  
> $\color{blue}{每个全局最大值都是局部最大值}$  
>***
> $\color{red}{极值定理}$ :  
>> 第 5 章中的 最大值小值定理: 连续函数在一个闭区间 [a, b] 内一定有一个全局最大值和一个全局最小值。如果函数不是连续的，或者尽管连续但定义域不是闭区间，这时该函数可能没有全局最大值或最小值。$\color{red}{这个定理的问题是没告诉我们最大值和最小值出现的位置}$  
>>  
>> 如果函数在 x = c 处的导数为零或导数不存在，我们就称 x = c 为$\color{red}{临界点}$  
>>  
>> $\color{red}{极值定理}$ : 假设函数 f 定义在开区间 (a, b) 内，并且点 c 在 (a, b) 区间内。如果点 c 为函数的局部最大值或最小值，那么点 c 一定为该函数的临界点。也就是说，$f^\prime(x) = 0$ 或 $f^\prime(x)$ 不存在。  
>***
> $\color{blue}{对于开区间相对极值只可能出现在临界点，对于闭区间相对极值可能出现在临界点或端点处。（注意: 临界点不一定是极值点）}$  
>***
> 求全局最大值和最小值步骤:  
>>$
（1）找出 f^\prime(x) ，并列出在 (a, b) 中 f^\prime(x) 不存在或 f^\prime(x) = 0 的点。也就是说，列出在开区间 (a, b) 内所有临界点。 \\
（2）把端点 x = a 和 x = b 也放入上述列表。 \\
（3）对于上述列表中每个点，将它们代入 f = f(x) 以求出它们所对应的函数值。 \\
（4）找出最大的函数值以及它所对应的 x 的值，这就是全局最大值。 \\
（5）用同样的方法找到最小的函数值和全局最小值。 \\
\\
无限定义域的情况略复杂，把无穷的情况也得考虑到极值的可能点中而它对应的函数值则得用极限来处理，但就算这个极限存在它 \\
既不是函数的临界点也不是函数的极值点（它甚至都不是个函数上的点，放入可能列表只是为了判断其余可能点是不是极值时作为参考值）。
$

**罗尔定理**
> 假设函数 f 在闭区间 [a, b] 上连续，在开区间 (a, b) 内可导。如果 f(a) = f(b) ，那么在开区间 (a, b) 内至少存在一点 c ，使得 $f^\prime(x) = 0$

**中值定理**
> 假设函数 f 在闭区间 [a, b] 内连续，在开区间 (a, b) 内可导，那么在区间 (a, b) 内至少有一点 c 使得 $f^\prime(c) = \frac{f(b) - f(a)}{b - a}$ ，如果以位移、速度、时间的关系来看，这里 $f^\prime(c)$ 表达的时瞬时速度，$\frac{f(b) - f(a)}{b - a}$ 表达的是平均速度。  
>***
> 中值定理的一些推导:
> 1. 如果对于在定义域 (a, b) 内的所有 x ，都有 $f^\prime(x) = 0$ ，那么函数 f 在开区间 (a, b) 内为常数函数。（证明导数处处为 0 的函数是常数函数）  
>  
> 2. 如果对于任意实数 x 都有 $f^\prime(x) = g^\prime(x)$ ，那么有 f(x) = g(x) + C（C为常数）。令 h(x) = f(x) - g(x) ，两边同时求导 $h^\prime(x) = f^\prime(x) - g^\prime(x) = 0$ ，所以 h(x) = C（C为常数）。所以 f(x) - g(x) = C 。  
>  
> 3. 如果函数 f 的导数始终为正，那么该函数为增函数（如果函数 f 的导数始终为负，那么该函数为减函数）。$
已知 f^\prime(x) > 0 ，令 b > a 则，根据中值定理开区间 (a, b) 内存在一个常数 c 使得 f^\prime(c) = \frac{f(b) - f(a)}{b - a} \quad\Rightarrow\quad f(b) - f(a) = f^\prime(c)(b - a) ，其中 f^\prime(c) > 0 且 b - a > 0 所以 f(b) - f(a) > 0  \color{red}{（这就证明了增函数，减函数的同理证明即可！！！！！）}
$

**二阶导数和图像**
> 假如知道对于开区间 (a, b) 内的所有 x ，$f^{\prime\prime}(x)$ > 0 。这意味这函数的一阶导数是个增函数，这在函数的图像上体现的就是函数图像在该开区间内是凹向上的。  
> 又假如知道对于开区间 (a, b) 内的所有 x ，$f^{\prime\prime}(x)$ < 0 。这意味这函数的一阶导数是个减函数，这在函数的图像上体现的就是函数图像在该开区间内是凹向下的。  
> 函数图像的这种凹性可能会发生变化，函数图像的凹性在凹向上和凹向下之间切换的点叫$\color{red}{拐点}$（这个表达有些不严谨！！！！！）  
>***
> 拐点:  
>> $
\color{blue}{如果 x = c 点是函数的拐点，则有 f^{\prime\prime}(c) = 0 。反过来 f^{\prime\prime}(c) = 0 ，则 x = c 就不一定是函数的拐点。} \\
例如 f(x) = x^4 ，x = 0 处 f^{\prime\prime}(0) = 0 ，但是从图像看 x = 0 的附近函数的图像一直凹向上的所以 x = 0 不是拐点。 \\
* 函数拐点的两侧 f^{\prime\prime}(x) 的符号一定是相反的。 \\
* 图像上拐点的特点是该点处的切线会穿过函数图像。\color{red}{拐点的一侧切线在曲线的上面，另一侧切线在曲线的下面。}
$  

**对于导数为零点的分类**
> $
对函数 f(x) 在 c 点处 f^\prime(c) = 0 说明 x = c 点处的切线是水平的。其实这个是一阶导数为零的位移直观的所得，那从它可以得到哪些信息呢？ \\
（1）x = c 可能为局部最大值点。 \\
（2）x = c 可能为局部最小值点。 \\
（3）x = c 可能为水平拐点。\\
那么怎么确定是这 3 个中的哪种？ \\
\color{blue}{
1. 用一阶导数（观察“x = c 附近”的一阶导数的符号）: \\
--- 首先，f^\prime(c) = 0 \\
--- 如果从左往右通过 c 点，f^\prime(x) 的符号由正变负，那么 c 点为局部最大值。 \\
--- 如果从左往右通过 c 点，f^\prime(x) 的符号由负变正，那么 c 点为局部最小值。 \\
--- 如果从左往右通过 c 点，f^\prime(x) 的符号不发生变化，那么 c 点为水平拐点。 \\
2. 用二阶导数（观察“x = c 处”的二阶导数的符号） \\
--- 首先，f^\prime(c) = 0 \\
--- 如果 f^{\prime\prime}(x) < 0 ，那么 c 点为局部最大值。 \\
--- 如果 f^{\prime\prime}(x) > 0 ，那么 c 点为局部最小值。 \\
--- 如果 f^{\prime\prime}(x) = 0 ，那么无法判定是哪种情况。（就得用上一个方法了）
}$  
>***
> $\color{blue}{临界点可能是极值点(局部最大值或局部最小值点)，或者是水平拐点}$
